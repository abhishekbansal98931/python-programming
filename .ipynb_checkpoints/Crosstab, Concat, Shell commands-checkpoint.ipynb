```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "country": ['India', 'India', 'India', 'India', 'Australia', 'Australia', 'Australia', 'SA', 'SA', 'India', 'India'],
    "state" : ['MP', 'MP', 'MP', 'UP', 'NSW', 'NSW', 'Victoria', 'Gauteng', 'Western Cape', 'Delhi','UP'],
    "city": ['Indore', 'Bhopal', 'Jabalpur', 'Lucknow', 'Sydney', 'Newcastle', 'Melbourne', 'Johannesburg', 'Cape Town', 'Delhi', 'Noida'],
    "sales": [150, 200, 100, 100, 200, 250, 300, 300, 200, 100, 50]
})
```

```python
df
```

```python
pd.crosstab(df['country'],df['state'], values=df['sales'], aggfunc=sum)
```

```python
df1 = df[df.country == 'India'].reset_index(drop=True)
df2 = df[df.country == 'Australia'].reset_index(drop=True)
df3 = df[df.country == 'SA'].reset_index(drop=True)
```

### Concat vertically (one below the other)

```python
#combining multiple dataframes by rows (basically union)
#pd.concat([df1,df2,df3])
new_df = pd.concat([df1,df2,df3], keys=['IND','AUS','SA'])
new_df
```

```python
#Retrieving specific keys from the dataframe created above
new_df.loc['IND']
```

### Concat horizontally (sideways)

```python
df1 = pd.DataFrame({'A': ['A0','A1','A2','A3']
                   ,'B': ['B0','B1','B2','B3']
                   ,'C': ['C0','C1','C2','C3']
                   ,'D': ['D0','D1','D2','D3']},
                  index=[0,1,2,3])

df2 = pd.DataFrame({'B': ['B0','B1','B2','B3']
                   ,'D': ['D0','D1','D2','D3']
                   ,'F': ['F0','F1','F2','F3']},
                  index=[2,3,6,7])
```

```python
#by default flex-direction is set as column, so we can change it to row
from IPython.display import display, HTML

CSS = """
.output {
    flex-direction: row;
}
"""

HTML('<style>{}</style>'.format(CSS))
```

```python
display(df1) 
display(df2)
```

```python
#join horizontally by index
#Outer join - default option (you don't need to specify)
pd.concat([df1,df2], axis=1, sort=False)
```

```python
#Inner join
pd.concat([df1,df2], axis=1, sort=False, join='inner')
```

```python
#Concat vertically but inner join
pd.concat([df1,df2], axis=0, sort=False, join='inner')
```
