```python
#Rank over partition by
df['rank'] = df.groupby(['country'])['sales'].rank(ascending = False)
```

```python
#Row number over partition by
df['rank'] = df.sort_values(['sales'], ascending = False).groupby(['country'])['sales'].cumcount() + 1
```

```python
#Dense rank over partition by
df['rank_dense'] = df.groupby(['country'])['sales'].rank(ascending = False, method = 'dense')
```

```python
#Custom filter

def custom_filter(x):
    if x['country'] == 'India':
        if x['sales'] > 100:
            return True
        else:
            return False
    else:
        return False

df[df.apply(custom_filter, axis = 1)]
#In apply function, axis = 1 will apply the function row wise and axis = 0 (default) will apply the function column wise
```

```python
#Alternatively
df[(df['country'] == 'India') & (df['sales'] > 100)]
```

### Grouped Aggregation with .groupby and .agg

```python
df_merged.groupby(['country','state']).agg({'sales_x':'sum'})
```

```python
filter_logic = (df_merged['country'] == 'India') | (df_merged['state'] == 'Australia') | (df_merged['sales_x'] > 0)

#calculate % of total sales in country, state
def percent_in_cs(x):
    return 100 * (x/x.sum())

df_merged[filter_logic].groupby(['country','state']).agg({'sales_x':'sum'})\
                       .groupby(level=0).apply(percent_in_cs)
#First groupby indexes rows by country and state. This is referred to as multiindexing. When chaining groupby statements 
```

```python
df_new = df.copy()
df_new.index = pd.MultiIndex.from_frame(df[['country','state','city']])
df_new = df_new.drop(columns=['country','state','city'])

#OR

df_new.set_index(['country','state','city'], inplace=True)

df_new.groupby(level=0).apply(lambda x: x/x.sum()).reset_index()
```

```python
#creating a date range
start_date = '2020-01-01'
end_date = '2020-05-31'
dates = pd.date_range(start = start_date, end = end_date, freq = '7D') #7D denotes separated by 7 days (weekly)
```

```python
dates
```

```python
#for each store we want these dates
store_list = [i+1 for i in range(10)]

date_list = [dates for i in range(0,len(store_list))]
```

```python
base_df = pd.DataFrame({'Store': store_list, 'Date': date_list})\
            .explode('Date')\
            .reset_index(drop=True)
#explode will pivot the date list in each row and create a new row for each date value
```

```python
base_df
```

```python
base_df['sales'] = 100
```

```python
#Sales of each store by month
base_df['Month'] = base_df['Date'].dt.month

sum_sales_monthly = base_df\
                        .groupby(['Store','Month'])\
                        .sum().reset_index()
```

```python
sum_sales_monthly.head(5)
```

```python
#4 month Rolling sales

sum_sales_monthly['monthly_sales'] = sum_sales_monthly.sort_values(['Store','Month'])\
                                    .groupby(['Store'])['sales']\
                                    .rolling(4, min_periods=1)\
                                    .sum()\
                                    .fillna(0)\
                                    .reset_index(drop = True)

#as_index = False will not make this as index
```

```python
sum_sales_monthly.head(5)
```

```python
#help(pd.DataFrame.rolling)
```

```python
#Transpose: converts rows to columns and columns to rows
sum_sales_monthly.describe().T
```

```python
#creating bins for sales
#using function cut
#lower bound exclusive and upper band inclusive
sum_sales_monthly['sales_bands'] = pd.cut(sum_sales_monthly['monthly_sales']
                                      , bins=[499,900,1300,1800]
                                      , labels=['Low','Medium','High'])
#bins can also be scalar for eg. bins = 5. In this case it will return 5 bins with 1/5th of the observations each
```

```python
sum_sales_monthly.head(5)
```

```python
sum_sales_monthly.groupby(['sales_bands']).agg({'monthly_sales': [min, max, len]})
```

### Pivot table and melt

```python
long = pd.DataFrame({'Cohorts': [1,1,1,2,2],
                     'Metric': ['M1','M2','M3','M1','M2'],
                     'Val' : [100,150,125,200,110]
                    })
```

```python
long
```

```python
wide = pd.pivot_table(data = long,
                      index = ['Cohorts'],
                      columns = 'Metric',
                      values = 'Val',
                      aggfunc = 'sum'
                     )
#aggfunct is optional in this specific case since there are no multiple values for the different combinations
#margins = True argument can also be added and it provides row and column totals
```

```python
wide
```

```python
wide2 = wide.reset_index()
```

```python
wide2
```

```python
long2 = pd.melt(wide2,
                id_vars=['Cohorts'],
                value_vars=['M1','M2','M3'],
                var_name='Metric',
                value_name='Val'
                        )
```

```python
long2
```
